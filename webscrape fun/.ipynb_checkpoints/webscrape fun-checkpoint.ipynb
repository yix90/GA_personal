```python
def get_title_youtube(url):
    html = urllib.urlopen(url).read()
    soup = BeautifulSoup(html, 'html.parser', from_encoding="utf-8")
    title = re.match(r'<.+>(.*)\s-.*<.+>', str(soup.title)).group(1)
    return title
```

```python
get_title_youtube('https://www.youtube.com/watch?v=fHiGbolFFGw')
```
