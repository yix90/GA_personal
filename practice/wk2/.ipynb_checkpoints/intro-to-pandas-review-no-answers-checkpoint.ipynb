**12. How many distinct districts are there?**

```python
print set(sanfran['PdDistrict'])
print len(set(sanfran['PdDistrict']))
```

**13. Which day of the week has the most crime?**

```python
d ={}
for day in set(sanfran['DayOfWeek']):
    d[day] = len(sanfran[sanfran['DayOfWeek']==day])
print max(d, key=d.get)
    
```

#### 14. Make a new DataFrame featuring the crime categories and the number of crimes per category.

```python
cat = {}
for crime in set(sanfran['Category']):
    cat[crime] = len(sanfran[sanfran['Category']==crime])
cat_d = pd.DataFrame({'Category':cat.keys(), 'No_Crimes':cat.values()})
cat_d
```

**15. Make a DataFrame that includes the districts and crime counts per district. Which district has the most crime?**

*Hint: You can use the `.sort_values()` function to sort your DataFrame by column.*

```python
dis = {}
for dist in set(sanfran['PdDistrict']):
    dis[dist] = len(sanfran[sanfran['PdDistrict']==dist])
cat_d = pd.DataFrame({'District':dis.keys(), 'No_Crimes':dis.values()})
cat_d.sort_values('No_Crimes', ascending=False)
```
