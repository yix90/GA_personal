```python
part = mbti.iloc[[0,1]]
part.head()
```

```python
def href_extract(x):
    lis=[]
    for line in x[2]:
        lin = line.split()
        for l in lin:
            if re.search(r'h.+youtube\S+', l):
                lis.append(re.search(r'h.+youtube\S+', l).group(0))
    return lis
```

```python
part.apply(href_extract, axis=1)[1]
```

```python
part.iloc[1,2]
```

```python
def img_extract(x):
    lis=[]
    for line in x[2]:
        lin = line.split()
        for l in lin:
            if l in ['.png','.jpg','.gif']:
                lis.append(re.search(r'http.+\.', l).group(0))
    return lis
```

```python
part.apply(img_extract, axis=1)[0]
```
