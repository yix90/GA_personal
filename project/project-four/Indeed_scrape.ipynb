```python
def diggy(tester):
    """
    Input part of HTML containing an indeed job post
    This function will dig out the necessary features 
    
    """
    nice_dict ={}
        
    #Title (sponsored)
    jobtitle1 = tester.find('b', class_='jobtitle')
    jobtitle2 = tester.find('h2', class_='jobtitle')
    if jobtitle1:
        nice_dict['title'] = jobtitle1.text
    elif jobtitle2:
        nice_dict['title'] = jobtitle2.text.strip()
    else:
        nice_dict['title'] = None
        
    #Company name
    compname = tester.find('span', class_='company')
    if compname:
        nice_dict['company'] = compname.text.strip()
    else:
        nice_dict['company'] = None

    #Rating score
    rating = tester.find('span', class_='rating')
    if rating:
        nice_dict['rating'] = rating['style'][6:10]
    else:
        nice_dict['rating'] = None

    #No. Reviews
    review = tester.find('span', class_='slNoUnderline')
    if review:
        nice_dict['review'] = review.text[:-7]
    else:
        nice_dict['review'] = None

    #Location
    location = tester.find('span', class_='location')
    if location:
        nice_dict['location'] = location.text
    else:
        nice_dict['location'] = None

#     #Sponsor (only the advertised job ad are sponsored)
#     sponsor = tester.find('span', class_='sponsoredGray')
#     if sponsor:
#         nice_dict['sponsor'] = sponsor.text
#     else:
#         nice_dict['sponsor'] = None
    
    #Salary info
    salary = tester.find_all('span', class_='no-wrap')
    if salary:
        nice_dict['salary'] = [s.text.strip() for s in salary]
    else:
        nice_dict['salary'] = None

    #Duration
    duration = tester.find('span', class_='date')
    if duration:
        nice_dict['duration'] = duration.text
    else:
        nice_dict['duration'] = None
        
    #Source
    source = tester.find('span', class_='sdn')
    if source:
        nice_dict['source'] = source.text
    else:
        nice_dict['source'] = None
        
    #Print summary
    summ = tester.find('span', id='job_summary')
    if summ:
        nice_dict['summary'] = summ.text.strip()
    else:
        nice_dict['summary'] = None
    return nice_dict
```

```python
driver = selenium.webdriver.Chrome('chromedriver.exe')
number = 0
df = pd.DataFrame()
driver.get("https://www.indeed.com.sg/m/jobs?q=data+science&l=Singapore")
while number < 1: 
    sleep(1)
    html = BeautifulSoup(driver.page_source, 'html.parser')
    for line in html.find_all('h2', class_='jobTitle'):
        addurl = line.find('a')['href']
        driver.get("https://www.indeed.com.sg/"+addurl)
        delishsoup = BeautifulSoup(driver.page_source, 'html.parser')
        jobdict = diggy(delishsoup)
        df = df.append(jobdict, ignore_index=True)
    for line in html.find_all('p'):
        if line.find('a'):
            number = 5
            for liner in line.find_all('a'):
                if liner.text == 'Next':
                    driver.get("https://www.indeed.com.sg/m/"+liner['href'])
                    number = 0
                    break
            if number == 0:
                break
driver.close()
```

```python
url = 'https://www.techinasia.com/jobs/search?country_name[]=Singapore&job_category_name[]=Data%20%26%20Analytics'
driver = selenium.webdriver.Chrome('chromedriver.exe')
driver.get(url)
sleep(10)
tomatosoup = BeautifulSoup(driver.page_source, 'html.parser')
tia_df = pd.DataFrame()
for arti in tomatosoup.find_all('article', class_='jsx-1346842819 card'):
    tia_dict={}
    tia_url = arti.find('a')['href']
    driver.get('https://www.techinasia.com'+tia_url)
    sleep(5)
    mushroomsoup = BeautifulSoup(driver.page_source, 'html.parser')
    garlicbread =  mushroomsoup.find('div', class_='jsx-4045607726 header-content')
    title = garlicbread.find('h1', class_='jsx-4045607726')
    if title:
        tia_dict['title'] = title.text
    else:
        tia_dict['title'] = None
    crumb = garlicbread.find_all('div', class_='jsx-4045607726')
    tia_dict['company'] = crumb[0].text
    tia_dict['location'] = crumb[1].text
    salary = garlicbread.find('div', class_='jsx-4045607726 compensation')
    if salary:
        tia_dict['salary'] = salary.text
    else:
        tia_dict['salary'] = None
    tia_dict['date'] = garlicbread.find('span', class_='jsx-4045607726 dates__created').text.strip()
    goulash = mushroomsoup.find_all('section', class_='jsx-4029956070')
    tia_dict['chunk'] = goulash[0].text.strip()
    tia_dict['skills'] = [mash.text for mash in goulash[1].find_all('span')]
    tia_df = tia_df.append(tia_dict, ignore_index=True)
    
tia_df.head()
```
