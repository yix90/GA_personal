```python
def try_mod(X,y):
    X_train = X[hous['YrSold'] < 2010]
    y_train = y[hous['YrSold'] < 2010]
    X_test = X[hous['YrSold'] >= 2010]
    y_test = y[hous['YrSold'] >= 2010]
    
    print "Start with Ridge"

    optimal_ridge = RidgeCV(alphas=np.logspace(0, 5, 200), cv=100)
    optimal_ridge.fit(X_train, y_train)
    print "Ridge Alpha:", optimal_ridge.alpha_
    ridge = Ridge(alpha = optimal_ridge.alpha_)
    ridge_scores = cross_val_score(ridge, X_train, y_train, cv=10)
    print "Cross validation scores:", ridge_scores
    print "Average score:", np.mean(ridge_scores)
    modelr = ridge.fit(X_train, y_train)
    yhat_r = modelr.predict(X_test)
    print "Prediction score:", modelr.score(X_test, y_test)
    
    print "_____________________________________________________________________________________________"
    print "Proceeding to Lasso"
        
    optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
    optimal_lasso.fit(X_train, y_train)
    print "Lasso Alpha:", optimal_lasso.alpha_
    lasso = Lasso(alpha=optimal_lasso.alpha_)
    lasso_scores = cross_val_score(lasso, X_train, y_train, cv=10)
    print "Cross validation scores:", lasso_scores
    print "Average score:", np.mean(lasso_scores)
    modell = lasso.fit(X_train, y_train)
    yhat_l = modell.predict(X_test)
    print "Prediction score:", modell.score(X_test, y_test)
    
    print "_____________________________________________________________________________________________"
    print "Last but not least, Elastic Net"
    
    l1_ratios = np.linspace(0.01, 1.0, 25)
    optimal_enet = ElasticNetCV(l1_ratio=l1_ratios, n_alphas=100, cv=10,verbose=1)
    optimal_enet.fit(X_train, y_train)
    print "Elastic Net Alpha:", optimal_enet.alpha_
    print "l1 ratio:", optimal_enet.l1_ratio_
    enet = ElasticNet(alpha=optimal_enet.alpha_, l1_ratio=optimal_enet.l1_ratio_)
    enet_scores = cross_val_score(enet, X_train, y_train, cv=10)
    print "Cross validation scores:", enet_scores
    print "Average score:", np.mean(enet_scores)
    modele = enet.fit(X_train, y_train)
    yhat_e = modele.predict(X_test)
    print "Prediction score:", modele.score(X_test, y_test)
    
    return ridge, lasso, enet
```

```python
from sklearn.preprocessing import binarize
def manual_confusion(ohat_pp, o_test):
    holddict = {'tp':[],'fp':[],'fn':[],'tn':[], 'thresh':[]}
    for i in np.arange(0.4,0.01,-0.01):
        newhat = binarize(ohat_pp, i)[:,1]
        #ot = o_test.ravel()
        holddict['tp'].append(np.sum((o_test == 1)&(newhat == 1)))
        holddict['fp'].append(np.sum((o_test == 0)&(newhat == 1)))
        holddict['tn'].append(np.sum((o_test == 0)&(newhat == 0)))
        holddict['fn'].append(np.sum((o_test == 1)&(newhat == 0)))
        holddict['thresh'].append(i)
    df = pd.DataFrame(holddict, columns=['thresh','tp','tn','fp','fn'])
    df['accuracy'] = (df['tp']+df['tn'])/(df.iloc[:,1:].apply(np.sum, axis=1))
    return df
```

```python
def manualer_confusion(ohat_pp, o_test, start, end, step):
    holddict = {'tp':[],'fp':[],'fn':[],'tn':[], 'thresh':[]}
    for i in np.arange(start, end, step):
        newhat = binarize(ohat_pp, i)[:,1]
        #ot = o_test.ravel()
        holddict['tp'].append(np.sum((o_test == 1)&(newhat == 1)))
        holddict['fp'].append(np.sum((o_test == 0)&(newhat == 1)))
        holddict['tn'].append(np.sum((o_test == 0)&(newhat == 0)))
        holddict['fn'].append(np.sum((o_test == 1)&(newhat == 0)))
        holddict['thresh'].append(i)
    df = pd.DataFrame(holddict, columns=['thresh','tp','tn','fp','fn'])
    df['accuracy'] = (df['tp']+df['tn'])/(df.iloc[:,1:].apply(np.sum, axis=1))
    return df
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split, cross_val_predict, cross_val_score
from sklearn.preprocessing import StandardScaler, binarize
from sklearn.metrics import classification_report, confusion_matrix

def try_logregcv(X,y,penalty='l1',cv=5, stand=False):
    pen_dict={'l1':'Lasso','l2':'Ridge'}
    logregcv = LogisticRegressionCV(penalty=penalty, solver='liblinear', scoring='roc_auc', cv=cv)
    stanfit = ss.fit(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
    if stand == True:
        X_train = stanfit.transform(X_train)
        X_test = stanfit.transform(X_test)
    model = logregcv.fit(X_train,y_train)
    yhat = model.predict(X_test)
    yhat_pp = model.predict_proba(X_test)
    yhat_ppdf = pd.DataFrame(yhat_pp, columns=['prob_0','prob_1'])
    score = model.score(X_test, y_test)
    if score < 0.9338:                                 #Stay true to our baseline accuracy!
        return 0
    print pen_dict[penalty]+" Score:", score, "k=",cv
    print pd.DataFrame(confusion_matrix(y_test, yhat))
    print yhat_ppdf.head()
    print classification_report(y_test, yhat)
    #conf_df = manual_confusion(yhat_pp, y_test)
    plotsort = yhat_ppdf['prob_1'].sort_values(ascending=False).reset_index(drop=True)
    plt.plot(plotsort)
    plt.title("Probability of abnormality (sorted)")
    plt.show()
```
