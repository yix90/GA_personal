Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).

Print out the pokemon power for each of your players.

````python
def power_cal(players, pokedex, player_id):
    power_l = []
    player_dic = players[player_id]
    for pokeind in player_dic['player_pokemon']:
        for attr in pokedex[pokeind].keys():
            if attr in ['name', 'type']:
                continue
            else:
                power_l.append(pokedex[pokeind][attr])
    print players[player_id]['player_name']+"'s power is "+str(sum(power_l))

        
power_cal(players,pokedex_ind, 2)    
````

## 9. Write a function to generate the full pokedex

---

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. The `PokedexNumber` should be used as the `pokemon_id` key values for the dictionary of pokemon.

Your function should:

1. Take the parsed pokedex information you created above as an argument.
2. Return a dictionary in the same format as your original pokedex you created before containing the information from the parsed full pokedex file.

To test the function, print out the pokemon with id = 100.

````python
def poke_dict(pokedex, pokemonid=None):
    pokedict = {}
    pokedict_refine = {}
    pokeselect = {}
    header = pokedex[0]
    for line in pokedex[1:]:
        pokedict_inner={}
        for i in range(len(line)):
            if header[i] == 'PokedexNumber':
                continue
            pokedict_inner[header[i]]=line[i]
            
#There are some Pokemon who share the same id but have different attributes,
#so for each identical id we add 0.1 to differentiate them in the dictionary.
        while line[0] in pokedict.keys():
            line[0] += 0.1

#While adding in Pokemon variations of the same id the alternative id went haywire e.g. 479.30000000000007
#which made no sense on display, but these values need to be kept for searching purposes
#So we have one for searching, and one for display.
        pokedict[line[0]]=pokedict_inner
        pokedict_refine[round(line[0],1)]=pokedict_inner
    if type(pokemonid) == int and pokemonid > 0 and pokemonid < 722:
        while pokemonid in pokedict.keys():
            pokeselect[round(pokemonid,1)] = pokedict[pokemonid]
            pokemonid += 0.1
        return pokeselect
    else:
        return pokedict_refine
        
        
        
poke_dict(pokedex_list, 6)
````

````python
def list_check(a, b):
    '''
    list_check(list_1, list_2)
    Check if items in list_1 is in list_2.
    '''
    for i in a:
       if i in b:
          return True

def attr_list(head, value):
    '''
    attr_list(head, value) where head = header or category/title listing and value = dictionary containing search parameters
    Returns a list in the order of the header while taking the values from the search parameters.
    If the parameters does not exist, the default will be 0.
    The list returns in this order, ['Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']
    
    For example,
    value = {'Attack':50, 'Speed':70, 'HP':90}
    return value = [0,90,50,0,0,0,70]
    '''
    attr_list=[]
    for hd in head[3:]:
        if hd in value.keys():
            attr_list.append(value[hd])
        else:
            attr_list.append(0)
    return attr_list

def attr_return(attr_l, head, body, typed=None):
    ret_list=[]
    for bod in body:
        out_dict={}
        if typed != None:                       #First check for type, if the type condition exists
            if typed == bod[2]:
                boo=True
            else:
                continue
        for l in range(len(attr_l)):
            if attr_l[l] > bod[l+3]:            #Next iterate through all the attributes
                boo=False
                break
            else:
                out_dict[head[l+3]]=bod[l+3]
                boo=True
        if boo == True:
            out_dict['Name']=bod[1]
            out_dict['Type']=bod[2]
            ret_list.append(out_dict)
    return ret_list
                
def name_filter(head, body, name):
    output_list=[]
    for poke in body:
        output_dict={}
        if name in poke:
            for i in range(len(poke)):
                output_dict[head[i]]=poke[i]
            output_list.append(output_dict)
    return output_list

#This is the main function
def search_poke(pokedex, filter_options):
    '''
    search_poke(pokedex, filter_options)
    Enter the filter_options as a dictionary, with the following information as such:
    {'Name':___, 'Type':_____, 'Attack'____, 'Defense':___, 'HP':____, 'SpecialAttack':___, 'SpecialDefense':____, 'Speed':___, 'Total':___}
    You may do so in any order. Please enter the value of Name and Type in string form, and the rest of the attributes as integers.
    
    If the Name attribute is included, the function will only return the results with the pokemon name as keyed, ignoring all other attribute.
    Otherwise the results will return the exact type for the Type value, and for other attributes pokemon with equal to or higher than stated in the filter_options.
    '''
    
    header = pokedex[0]
    body = pokedex[1:]
    if 'Name' in filter_options.keys():
        return name_filter(header, body, filter_options['Name'])
    elif list_check(filter_options.keys(), header[2:]) == True:
        attr_l = attr_list(header, filter_options)
        try:
            type_select = filter_options['Type']
        except:
            type_select=None
        return attr_return(attr_l, header, body, type_select)
    else:
        print "Your search returned no results."
        return None

search_poke(pokedex_list, {'Total':780,'Type':'DragonFlying'})
````
